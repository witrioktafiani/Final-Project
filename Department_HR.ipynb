## 4.3 Clustering Model

```python
#Mendefinisikan color palette
palette = ["#e9311a","#ed6335","#026c80","#8db4ad","#064c72"]
cmap = colors.ListedColormap(["#b9c4ca", "#eeae94", "#e9bc6d", "#ca5a4b", "#64262e", "#84a725"])
```

```python
#Membuat data copy
data_clusters = data_robust.copy()
```

### 1 Dimensionality Reduction

**Principal component analysis (PCA)** adalah sebuah teknik untuk mengurangi dimensi data, namun meminimalkan kehilangan informasi dari data tersebut.

```python
#PCA untuk reduce dimentions menjadi 3
pca = PCA(n_components=3)
pca.fit(data_clusters[num_col])
PCA_ds = pd.DataFrame(pca.transform(data_clusters[num_col]), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#3D of data Reduced Dimension
x = PCA_ds["col1"]
y = PCA_ds["col2"]
z = PCA_ds["col3"]
#Plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="blue", marker="o" )
ax.set_title("Plot 3D The Reduced Dimension")
plt.show()
```

Plot di atas adalah gambaran data setelah dilakukan dimensional reduction dengan Principal component analysis (PCA). Plot tersebut dapat dibagi menjadi beberapa cluster, untuk melihat pembagian clusternya maka digunakan Elbow Method

### 2 Elbow Method

```python
#Menghitung nilai dari k (number of clusters)
distortions = []
K = range(1,7)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(PCA_ds)
    distortions.append(kmeanModel.inertia_)
```

```python
#plot elbow graph
plt.figure(figsize=(15,5))
plt.rcParams.update({'axes.grid': True})
plt.plot(K, distortions, 'bx-', color=palette[-1])
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method')
sns.despine()
plt.show()
```

Untuk melihat lebih jelas nilai *k* maka akan digunakan KElbowVisualizer 

```python
#Elbow Method using KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=6)

visualizer.fit(PCA_ds)
visualizer.show()
```

Hasil yang didapatkan berdasarkan Elbow Method yaitu nilai K = 3, maka akan dibentuk 3 clusters

### 3 Silhouette Score

```python
#Silhouette score Elbow using KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,30),metric='silhouette', timings= True)
visualizer.fit(PCA_ds) 
visualizer.show() 
```

Berdasarkan Silhoutte Score sebesar 0.503 maka model cluster dengan k = 3 clusters merupakan model yang rata-rata.

### 4 Agglomerative Clustering Model

```python
#Agglomerative Clustering model 
AC = AgglomerativeClustering(n_clusters=3)

#fit model and predict clusters
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC

#menambahkan clusters ke dalam original dataframe
data_clusters["Clusters"]= yhat_AC
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = cmap)
ax.set_title("The Plot Of The Clusters")
plt.show()
```
